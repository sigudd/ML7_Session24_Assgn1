```python
listA = ['City1','City2','Avg Fare','Distance','Avg Weekly Pass','Mkt Leader','Mkt Share','Avg Fare','Low Price Air','Low Mkt Share','Price']
```

```python
afpd = pd.read_fwf('http://www.stat.ufl.edu/~winner/data/airq402.dat', sep='\s\s+', engine='python')
afpd.shape
```

```python
afpd.columns = ['City1','City2','Avg Fare','Distance','Avg Weekly Pass','Mkt Leader','Mkt Share','Mkt Avg Fare','Low Price Air','Low Mkt Share','Price']
```

```python
afpd.head()
```

```python
afpd.shape
# imports
import pandas as pd
import matplotlib.pyplot as plt

# this allows plots to appear directly in the notebook
%matplotlib inline
```

```python
fig, axs = plt.subplots(1, 3, sharey=True)
afpd.plot(kind='scatter', x='Avg Weekly Pass', y='Avg Fare', ax=axs[0], figsize=(16, 8))
afpd.plot(kind='scatter', x='Distance', y='Avg Fare', ax=axs[1])
afpd.plot(kind='scatter', x='Mkt Share', y='Avg Fare', ax=axs[2])
```

```python
afpd.head()
```

```python
from sklearn.model_selection import train_test_split

y = afpd[["Avg Fare"]]
X = afpd.drop(['City1','City2','Mkt Leader','Low Price Air','Price','Avg Fare','Mkt Avg Fare'],axis=1)

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
import seaborn as sns


# follow the usual sklearn pattern: import, instantiate, fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python

# print intercept and coefficients
print(lm.intercept_)
print('*****')
print(lm.coef_)
print('***')
print(lm.score(X_train, y_train))
```

```python
print(lm.score(X_test, y_test))
```
